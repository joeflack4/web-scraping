```python
# Import stuff
import requests
from bs4 import BeautifulSoup
url = 'http://www.slate.com/topics/s/slate_book_review.html'
# Grab data from a website
response = requests.get(url)
response
```

```python
# Let's scrape :3
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Let's look for all the titles! (at this website, titles use the class 'hed')
titles = soup.find_all('span', class_='hed')
titles
```

```python
# List comprehension
titles_list =[title.get_text() for title in titles]

# Imperative coding style (like JavaScript)
non_comprehension_title_list = []
for title in titles:
    non_comprehension_title_list.append(title.get_text())

# They're the same!
print(non_comprehension_title_list == titles_list)
```
